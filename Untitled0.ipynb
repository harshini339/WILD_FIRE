```python
import os
import shutil

# 🔹 INPUT DATASET PATH
base_input = "/content/drive/MyDrive/wildfire/forest"

# 🔹 OUTPUT PATH
base_output = "/content/drive/MyDrive/wildfire/wildfire_yolo3"

splits = ["train", "val", "test"]

# 🔹 Class mapping
classes = {
    "fire": 0,
    "nofire": 1,
    "smoke": 2,
    "smokefire": 3
}

for split in splits:
    input_path = os.path.join(base_input, split)

    images_out = os.path.join(base_output, split, "images")
    labels_out = os.path.join(base_output, split, "labels")

    os.makedirs(images_out, exist_ok=True)
    os.makedirs(labels_out, exist_ok=True)

    img_count = 0

    for class_name in classes:
        class_folder = os.path.join(input_path, class_name)

        if not os.path.exists(class_folder):
            continue

        for img_file in os.listdir(class_folder):
            if not img_file.lower().endswith((".jpg", ".jpeg", ".png")):
                continue

            img_path = os.path.join(class_folder, img_file)

            new_name = f"{split}_{img_count}.jpg"

            # Copy image
            shutil.copy(img_path, os.path.join(images_out, new_name))

            # Create label
            label_path = os.path.join(labels_out, f"{split}_{img_count}.txt")
            class_id = classes[class_name]

            with open(label_path, "w") as f:
                f.write(f"{class_id} 0.5 0.5 1.0 1.0\n")

            img_count += 1

    print(f"✅ {split} done!")

print("🔥 FULL YOLO dataset ready!")
```

```python
custom_yaml = """
# YOLOv11 Custom Model for Wildfire Detection
nc: 4

backbone:
  - [-1, 1, Conv,  [64, 3, 2]]
  - [-1, 1, Conv,  [128, 3, 2]]
  - [-1, 1, C2f,   [128, 2]]
  - [-1, 1, Conv,  [256, 3, 2]]
  - [-1, 1, C2f,   [256, 2]]
  - [-1, 1, Conv,  [512, 3, 2]]
  - [-1, 1, C2f,   [512, 2]]
  - [-1, 1, SPPF,  [512, 5]]

head:
  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]
  - [[-1, 4], 1, Concat, [1]]
  - [-1, 1, C2f, [256, 2]]

  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]
  - [[-1, 2], 1, Concat, [1]]
  - [-1, 1, C2f, [128, 2]]

  - [-1, 1, Conv, [256, 3, 2]]
  - [[-1, 10], 1, Concat, [1]]
  - [-1, 1, C2f, [256, 2]]

  - [-1, 1, Conv, [512, 3, 2]]
  - [[-1, 6], 1, Concat, [1]]
  - [-1, 1, C2f, [512, 2]]

  - [[13, 16, 19], 1, Detect, [nc]]
"""

with open("/content/drive/MyDrive/wildfire/custom_wildfire.yaml", "w") as f:
    f.write(custom_yaml)
```

```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/wildfire/custom_wildfire.yaml")

print(f"Model parameters: {sum(p.numel() for p in model.model.parameters() if p.requires_grad):,}")
print(model.model)

results = model.train(
    data="/content/drive/MyDrive/wildfire/wildfire.yaml",
    epochs=20,
    imgsz=416,
    batch=8,
    pretrained=False,
    optimizer='auto',
    lr0=0.001,
    lrf=0.01,
    momentum=0.937,
    weight_decay=0.0005,
    patience=15,
    seed=0,
    save=True,
    plots=True,
    workers=8,
    project="/content/drive/MyDrive/wildfire",
    name="yolo_results",
    exist_ok=True
)
```

```python
from ultralytics import YOLO

# Load model
model = YOLO("/content/drive/MyDrive/wildfire/yolo_results/weights/best.pt")

# Evaluate
metrics = model.val()

# Extract values
map50 = metrics.box.map50
map5095 = metrics.box.map
precision = metrics.box.mp
recall = metrics.box.mr

print("\n=== MODEL PERFORMANCE ===")
print(f"mAP@50: {map50*100:.2f}%")
print(f"mAP@50-95: {map5095*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
accuracy = (precision + recall) / 2
print(f"Approx Accuracy: {accuracy*100:.2f}%")

```

```python
from ultralytics import YOLO
import gradio as gr

# Load model
model = YOLO("/content/drive/MyDrive/wildfire/yolo_results/weights/best.pt")

def predict(image):
    # prediction
    results = model.predict(
        source=image,
        conf=0.5
    )

    # Annotated image (same as .show())
    annotated_img = results[0].plot()

    # Extract predictions
    boxes = results[0].boxes
    data = []

    if boxes is not None:
        for box in boxes:
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            class_name = model.names[cls_id]

            data.append([class_name, round(conf, 2)])

    # If nothing detected
    if len(data) == 0:
        data = [["No Detection", 0.0]]

    return annotated_img, data


# Gradio UI
demo = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="filepath"),
    outputs=[
        gr.Image(type="numpy", label="Detection Result"),
        gr.Dataframe(headers=["Class", "Confidence"])
    ],
    title="🔥 Wildfire Detection",
    description="Detect fire, smoke, smokefire using YOLO"
)

demo.launch()
```
